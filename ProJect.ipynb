```python
import datetime
import pyfiglet

# Making 4 Dictionaries 3 for products availability and one for the prices
Prices = {
    "Cola": 130,
    "Cola Zero": 130,
    "Lemonade": 100,
    "Sahlab": 90,
    "Airan": 90,
    "3 in 1": 120,
    "Ice Tea": 100,
    "Ice Cappucino": 150,
    "Ice Latte": 150,
    "Orange Juice": 75
}

Jableh_Availability = {
    "Cola": 50,
    "Cola Zero": 33,
    "Lemonade": 25,
    "Sahlab": 12,
    "Airan": 78,
    "3 in 1": 44,
    "Ice Tea": 56,
    "Ice Cappucino": 59,
    "Ice Latte": 45,
    "Orange Juice": 17
}

Tartous_Availability = {
    "Cola": 41,
    "Cola Zero": 23,
    "Lemonade": 27,
    "Sahlab": 45,
    "Airan": 12,
    "3 in 1": 33,
    "Ice Tea": 20,
    "Ice Cappucino": 70,
    "Ice Latte": 54,
    "Orange Juice": 39
}

Latakia_Availability = {
    "Cola": 40,
    "Cola Zero": 53,
    "Lemonade": 65,
    "Mors": 8,
    "Airan": 15,
    "Tarkhon": 36,
    "Ice Tea": 20,
    "Ice Cappucino": 14,
    "Ice Latte": 50,
    "Orange Juice": 23
}

# Centering width
terminal_width = 80

# Time Zone (GMT+3)
def get_time():
    tz = datetime.timezone(datetime.timedelta(hours=3))
    now = datetime.datetime.now(tz)
    return now.strftime("%Y-%m-%d   %H:%M:%S ")


# Display drinks with numbers
def show_drinks(store_dict):
    print("\nAvailable Drinks:")
    print("------------------------")
    for i, (drink, amount) in enumerate(store_dict.items(), start=1):
        print(f"{i}. {drink}: {amount} available | Price: {Prices[drink]}")


# Show all prices
def show_all_prices():
    print("\nAll Drinks and Prices:")
    print("------------------------")
    for drink, price in Prices.items():
        print(f"{drink}: {price}")


# store selection
def choose_store():
    print("\nChoose a store:")
    print("1. Jableh")
    print("2. Tartous")
    print("3. Latakia")

    store_choice = input("Enter store number: ")

    if store_choice == "1":
        return Jableh_Availability, "Jableh"
    elif store_choice == "2":
        return Tartous_Availability, "Tartous"
    elif store_choice == "3":
        return Latakia_Availability, "Latakia"
    else:
        print("Invalid choice.")
        return None, None


# Receipt with time and date
def print_receipt(name, address, store, drink, quantity, delivery, total):
    print("\n")
    print("RECEIPT".center(terminal_width))
    print(f"Date & Time: {get_time()}".center(terminal_width))
    print(f"Buyer: {name}".center(terminal_width))
    print(f"Address: {address}".center(terminal_width))
    print(f"Drink: {drink}".center(terminal_width))
    print(f"Quantity: {quantity}".center(terminal_width))
    print(f"Store: {store}".center(terminal_width))
    print(f"Delivery: {delivery}".center(terminal_width))
    print(f"Total: {total}".center(terminal_width))
    print("Thank you for your order!".center(terminal_width))
    print("\n")


# Welcome screen with pyfiglet
welcome_text = pyfiglet.figlet_format("WELCOME TO THE STORE", font="slant")
print(welcome_text)


# Main Program Loop
while True:
    print("\nAre you:")
    print("1. Employee")
    print("2. Buyer")
    print("0. Exit")

    job = input("Enter your choice: ")

    # EMPLOYEE
    if job == "1":
        password = input("Enter the password: ")
        if password == "11235813":
            print("\nAccess granted.".center(terminal_width))

            while True:
                print("\nEmployee Menu:")
                print("1. Add/Edit Products")
                print("2. Delete Products")
                print("0. Log Out")
                choice = input("Enter your choice: ")

                # Add/Edit products
                if choice == "1":
                    cnt = int(input("How many items do you want to add/edit? "))
                    store, store_name = choose_store()
                    if not store:
                        continue

                    for i in range(cnt):
                        key = input("Enter product name: ").title()
                        value = int(input("Enter quantity: "))
                        price = int(input("Enter price: "))
                        store[key] = value
                        Prices[key] = price

                    print("Products updated successfully!".center(terminal_width))

                # Delete product
                elif choice == "2":
                    store, store_name = choose_store()
                    if not store:
                        continue

                    key = input("Enter product to delete: ").title()
                    if key in store:
                        del store[key]
                        print(f"{key} deleted from {store_name} store.".center(terminal_width))
                    else:
                        print("Product not found.".center(terminal_width))

                elif choice == "0":
                    print("Logging out...")
                    break
                else:
                    print("Invalid choice.".center(terminal_width))

        else:
            print("Wrong password!".center(terminal_width))


    # BUYER
    elif job == "2":
        name = input("Enter your name: ")
        phone = input("Enter your phone number: ")

        print(pyfiglet.figlet_format(f"Welcome {name}", font="small"))

        while True:
            print("\nBuyer Menu:")
            print("1. Check Availability")
            print("2. Order Drinks")
            print("3. View All Drinks")
            print("0. Log Out")

            option = input("Enter your choice: ")

            if option == "1":
                store, store_name = choose_store()
                if store:
                    print(f"\nAvailability in {store_name}:")
                    show_drinks(store)

            elif option == "2":
                store, store_name = choose_store()
                if not store:
                    continue
                show_drinks(store)

                drink_list = list(store.keys())
                choice_num = int(input("\nEnter drink number: "))

                if choice_num < 1 or choice_num > len(drink_list):
                    print("Invalid selection.".center(terminal_width))
                    continue

                drink = drink_list[choice_num - 1]

                quantity = int(input("Enter the quantity: "))
                if quantity > store[drink]:
                    print("Not enough in stock.".center(terminal_width))
                    continue

                total = Prices[drink] * quantity
                print(f"Total price: {total}")

                delivery_option = input("Delivery or Pickup? (d/p): ").lower()
                if delivery_option == "d":
                    address = input("Enter delivery address: ")
                    delivery = "Home Delivery"
                else:
                    address = "Pickup"
                    delivery = "Pickup"

                store[drink] -= quantity

                print_receipt(name, address, store_name, drink, quantity, delivery, total)

            elif option == "3":
                show_all_prices()

            elif option == "0":
                print(pyfiglet.figlet_format("Goodbye!", font="slant"))
                break

            else:
                print("Invalid choice.".center(terminal_width))


    elif job == "0":
        print(pyfiglet.figlet_format("Goodbye!", font="slant"))
        break

    else:
        print("Invalid input.".center(terminal_width))
```
